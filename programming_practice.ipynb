```python
path='/Users/anca/Desktop/dev/primer/g2-onprem/file'
lines = open(path,'r').readlines()
print(lines)
```

```python
def append_number_matches(line, numbers_array):
    numbers_dict = {"one": "1", "two": "2", "three": "3", "four": "4", "five": "5", "six": "6", "seven": "7", "eight": "8", "nine": "9"}
    for number in numbers_dict:
        occurences = [i for i in range(len(line)) if line.startswith(number, i)]
        for occurence in occurences:
            numbers_array.append((numbers_dict[number], occurence))


def get_first_and_last_num(line):
    numbers = "0123456789"

    found_nums = []
    for idx in range(len(line)):
        if line[idx] in numbers:
            found_nums.append((line[idx], idx))
    append_number_matches(line, found_nums)

    # sort the list of tuples by the index of the number
    found_nums.sort(key=lambda x: x[1])
    # print(found_nums)
    if len(found_nums) == 0:
        return None, None
    return found_nums[0][0], found_nums[-1][0]
```

```python
# test number matching

# test that multiples of the same number are found
line ="oneoneone9"
nums = []
append_number_matches(line, nums)
assert nums == [('1', 0), ('1', 3), ('1', 6)]
assert get_first_and_last_num(line) == ('1', '9')

# test edge case of no numbers
line = "no numbers here"
assert get_first_and_last_num(line) == (None, None)

# test edge case of only one number
line = "eight"
assert get_first_and_last_num(line) == ('8', '8')
```

```python

sum = 0
# print(file)
for line in lines:
    # print(line)
    new_line = "".join(reversed(line))
    new_line = line[::-1]
    # print("new line:" , new_line)
    first_num, last_num = get_first_and_last_num(line)
    # print(first_num, last_num)
    two_digit = str(first_num )+ str(last_num)
    sum += int(two_digit)
print(sum)

```
